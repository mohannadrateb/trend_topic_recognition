```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd_ai = pd.read_csv('ai_merged.csv')
pd_retweet = pd_ai[['date','retweet_count']] 
pd_follower_count = pd_ai[['date','follower_count']]
pd_likes = pd_ai[['date','favorite_count']]
```

```python
pd_ai_count = pd.read_csv('ai_2020_count.csv')
```

```python
mean_retweeets =  pd_ai.mean()['retweet_count']
std_retweets = pd_ai.std()['retweet_count']
mean_follower_count =  pd_ai.mean()['follower_count']
std_follower_count = pd_ai.std()['follower_count']
mean_likes =  pd_ai.mean()['favorite_count']
std_likes = pd_ai.std()['favorite_count']
```

```python
mean_count_tweets = pd_ai_count.mean()['count']
std_count_tweets = pd_ai_count.std()['count']
```

```python
cut_off_retweets = std_retweets * 3
cut_off_follower_count = std_follower_count * 3
cut_off_likes = std_likes * 3
cut_off_count_tweets = std_count_tweets * 3
```

```python
lower_retweet, upper_retweet = mean_retweeets - cut_off_retweets,mean_retweeets + cut_off_retweets
lower_follower_count, upper_follower_count = mean_follower_count - cut_off_retweets,mean_follower_count + cut_off_retweets
lower_likes, upper_likes = mean_likes - cut_off_likes,mean_likes + cut_off_likes
lower_count,upper_cont = mean_count_tweets - cut_off_count_tweets,mean_count_tweets + cut_off_count_tweets






```

```python
pd_retweet.loc[(pd_retweet.retweet_count >upper_retweet) | (pd_retweet.retweet_count <lower_retweet), 'retweet_count' ] = mean_retweeets
pd_follower_count.loc[(pd_follower_count.follower_count >upper_follower_count) | (pd_follower_count.follower_count <lower_follower_count), 'follower_count' ] = mean_follower_count
pd_likes.loc[(pd_likes.favorite_count > upper_likes) | (pd_likes.favorite_count < lower_likes), 'favorite_count' ] = mean_likes
pd_ai_count.loc[(pd_ai_count['count'] >upper_cont) | (pd_ai_count['count'] <lower_count), 'count' ] = mean_count_tweets



```

```python
plt.rcParams['figure.figsize'] = [10, 8]
pd_retweet.plot(label ="retweet_ai")
plt.xlabel('day',fontweight='bold')
plt.ylabel('NUM_OF_RETWEETS',fontweight='bold')
```

```python
pd_follower_count.plot(label = 'follower_count')
plt.xlabel('day',fontweight='bold')
plt.ylabel('NUM_OF_FOLLLOWERS',fontweight='bold')
plt.legend()
```

```python
pd_likes.plot(label = 'number_of_likes')
plt.xlabel('day',fontweight='bold')
plt.ylabel('NUM_OF_LIKES',fontweight='bold')
plt.legend()
```
